```python
import pandas as pd

df = pd.read_csv('/content/Dataset .csv')
df.head()
```

```python
print(f"Shape of the DataFrame: {df.shape}")
df.info()
```

```python
df['City'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

df['City'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10).plot(kind='bar')

plt.title("Top Cities by Average Rating")
plt.ylabel("Average Rating")

plt.show()
```

```python
import seaborn as sns

sns.scatterplot(
    x='Longitude',
    y='Latitude',
    data=df
)

plt.title("Restaurant Locations")
plt.show()
```

```python
df.groupby('City')['Price range'].mean().sort_values(ascending=False).head(10)
```

Location-based analysis was performed using city, latitude, and longitude data. It was observed that some cities had a higher concentration of restaurants and better average ratings compared to others.
